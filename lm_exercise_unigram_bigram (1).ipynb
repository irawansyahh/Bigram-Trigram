**3.** Masukkan contoh paragraf yang terdiri atas beberapa kalimat, sebagai contoh teks yang akan dibangun model bahasa-nya.

```python
par = "Pesilat Indonesia kembali meraih medali emas ke-12, setelah Pipiet Kamelia berhasil menumbangkan pesilat Vietnam Thi Cam Nhi Nguyen di babak final pertandingan cabang olahraga Pencak Silat di ajang Asian Games 2018 . Pipiet menang telak 5-0 atas Thi Cam di kelas D putri 60kg-65 kg, yang berlangsung di Padepokan Pencak Silat Taman Mini Indonesia Indah (TMII), Jakarta Timur, Rabu petang. Dengan kemenangan ini, Pipiet berhasil membawa medali emas, sementara Thi Cam harus puas juara kedua dengan medali perak." 

print(par)
```

**4.** Ubah penulisan teks menjadi huruf kecil semua. Apa tujuannya?

```python
lc_par = par.lower()
print(lc_par)
```

**5.** Lakukan tokenisasi teks. Tokenisasi menjadi beberapa kalimat, lalu menjadi beberapa token.

```python
sent_text = nltk.sent_tokenize(lc_par) # tokenisasi menjadi beberapa kalimat
print(sent_text)

# loop per kalimat, tambahkan tag <s> di awal kalimat dan </s> di akhir kalimat
mod_sentences = [] # array/list untuk menampung semua kalimat teks awal
for sentence in sent_text:
    mod_sentence = []
    sent_tokens = nltk.word_tokenize(sentence)
    mod_sentence.append('<s>')
    for token in sent_tokens:
      mod_sentence.append(token)
    mod_sentence.append('</s>')
    mod_sentences.append(mod_sentence)


print(mod_sentences)
```

**6.** Masukkan data kata yang muncul pada teks beserta frekuensi kemunculannya ke dictionary freq_tab{}

```python
freq_tab = {}
total_count = 0
for sentence in mod_sentences:
  for token in sentence:
    if token in freq_tab:
        freq_tab[token] += 1 # kata sudah ada di dictionary, update frekuensinya
    else:
        freq_tab[token] = 1 # kata belum ada di dictionary 
    total_count += 1
        
print(freq_tab)
```

**7.** Buat model unigram: hitung probability unigram, simpan dalam dictionary prob_tab{}

```python
prob_tab = {}
for token in freq_tab:
    prob_tab[token] = freq_tab[token]/total_count
    
print(prob_tab)
```

**8.** Tes hitung probability sebuah kalimat berdasarkan model unigram yang sudah dibuat

```python
test_sentence = 'Pesilat Indonesia meraih emas'
lc_test_sentence = test_sentence.lower()
test_tokens = nltk.word_tokenize(lc_test_sentence)
total_prob = 1.0
for test_token in test_tokens:
    total_prob = total_prob * prob_tab[test_token]
    print(test_token)
    
print(total_prob)
```

**9.** Hitung frekuensi kemunculan bigram, masukkan dalam freq_bigram_tab. Key adalah pasangan unigram yang membentuk bigram

```python
freq_bigram_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])
    if curr_bigram in freq_bigram_tab:
      freq_bigram_tab[curr_bigram] += 1 # bigram sudah ada di dictionary, update frekuensinya
    else:
      freq_bigram_tab[curr_bigram] = 1 # bigram belum ada di dictionary
```

Periksa isi freq_bigram_tab

```python
print(freq_bigram_tab)
```

**10.** Pembuatan model probability bigram

```python
bigram_prob_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])

    if curr_bigram not in bigram_prob_tab:  
      bigram_prob_tab[curr_bigram] = freq_bigram_tab[curr_bigram]/freq_tab[sentence[i-1]] # bigram belum ada di dictionary, hitung probability  
 
print(bigram_prob_tab)
```

**11.** Buat fungsi untuk mencari kata selanjutnya yang mempunyai bigram prob tertinggi berdasar input kata sebelum

```python
def get_max_bigram_prob(prev_word):  
  max_prob_next_word = ''
  bigram_cand = {k: v for k, v in bigram_prob_tab.items() if k[0]==prev_word}
  if len(bigram_cand) > 0:
    bigram_cand = sorted(bigram_cand.items(), key=lambda x: x[1], reverse=True)
    k = bigram_cand[0]
    print('bigram dengan probability tertinggi = '+str(k))
    max_prob_next_word = k[0][1]
  
  return max_prob_next_word
```

**12.** Tes prediksi kata selanjutnya

```python
#prev_token = 'medali'
#prev_token = 'pesilat'
#prev_token = 'inggris'
prev_token = 'bola'
prev_token = '</s>'
predicted_next_word = get_max_bigram_prob(prev_token)
if len(predicted_next_word) > 0:
  print('prediksi kata selanjutnya setelah kata '+prev_token+': '+predicted_next_word)
else:
  print('kata sebelum tidak ditemukan di vocabulary atau tidak ada bigram yang mengandung kata tersebut sebagai previous token')
```
